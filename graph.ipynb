```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_syhw = pd.read_csv('./results_csv_sy/mvtec_metal_nut_syhw.csv')[['subname', 'iou']]
df_ssa = pd.read_csv('./results_csv_sy/mvtec_metal_nut_saa.csv')[['subname', 'iou']]
df_ramllama = pd.read_csv('./results_csv_sy/mvtec_metal_nut_ram_llama.csv')[['subname', 'iou']]
```

```python
iou_syhw = df_syhw['iou']
iou_saa = df_ssa['iou']
iou_ramllama = df_ramllama['iou']
```

```python
iou_syhw_lst = [float(iou_syhw[i][1:6]) for i in range(len(df_syhw))]
iou_saa_lst = [float(iou_saa[i][1:6] )for i in range(len(df_syhw))]
iou_ramllama_lst = [float(iou_ramllama[i][1:6]) for i in range(len(df_syhw))]
```

```python
acc_syhw_lst = [float(iou_syhw[i][9:14]) for i in range(len(df_syhw))]
acc_saa_lst = [float(iou_saa[i][9:14]) for i in range(len(df_syhw))]
acc_ramllama_lst = [float(iou_ramllama[i][9:14]) for i in range(len(df_syhw))]
```

```python
labels = list(df_syhw['subname'].values)
data1 = iou_syhw_lst
data2 = iou_ramllama_lst
data3 = iou_saa_lst

x = range(len(labels))  # x축 위치

# 막대그래프 그리기
fig, ax = plt.subplots(figsize=(8, 5))  # 그래프 사이즈 설정
bar_width = 0.25  # 막대 너비

bar1 = ax.bar(x, data1, width=bar_width, label='SYHW(ours)')
bar2 = ax.bar([i + bar_width for i in x], data2, width=bar_width, label='RAM+Llama')
bar3 = ax.bar([i + 2 * bar_width for i in x], data3, width=bar_width, label='SAA+')

# 막대 위에 값 표시
for bars in [bar1, bar2, bar3]:
    for rect in bars:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.3f}', ha='center', va='bottom')


# 라벨, 타이틀, 범례 추가
ax.set_xlabel('Labels')
ax.set_ylabel('Values')
ax.set_title('Bar Chart of IoU')
ax.set_xticks([i + bar_width for i in x])
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.0)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
labels = list(df_syhw['subname'].values)
data1 = acc_syhw_lst
data2 = acc_ramllama_lst
data3 = acc_saa_lst

x = range(len(labels))  # x축 위치

# 막대그래프 그리기
fig, ax = plt.subplots(figsize=(8, 5))  # 그래프 사이즈 설정
bar_width = 0.25  # 막대 너비

bar1 = ax.bar(x, data1, width=bar_width, label='SYHW(ours)')
bar2 = ax.bar([i + bar_width for i in x], data2, width=bar_width, label='RAM+Llama')
bar3 = ax.bar([i + 2 * bar_width for i in x], data3, width=bar_width, label='SAA+')

# 막대 위에 값 표시
for bars in [bar1, bar2, bar3]:
    for rect in bars:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.3f}', ha='center', va='bottom')


# 라벨, 타이틀, 범례 추가
ax.set_xlabel('Labels')
ax.set_ylabel('Values')
ax.set_title('Bar Chart of Accuracy')
ax.set_xticks([i + bar_width for i in x])
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.3)
ax.legend()

plt.tight_layout()
plt.show()

```
